#ü§ñ Modelado Predictivo

## Datos de entrenamiento y prueba

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=47)
```

## Generando un modelo de referencia - Baseline

```python
from sklearn.dummy import DummyClassifier
```

```python
dummy = DummyClassifier()
dummy.fit(X_train, y_train)
```

```python
#Evalua la performance del modelo dummy
dummy.score(X_test, y_test)
```

 ## Uso de √°rboles de decisi√≥n (Modelo de √°rbol)

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
modelo_arbol = DecisionTreeClassifier(random_state=47)
modelo_arbol.fit(X_train, y_train)
```

```python
modelo_arbol.score(X_test, y_test)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
nombre_columnas = X_train.columns
```

```python
plt.figure(figsize=(80,25))
plot_tree(modelo_arbol, filled=True, fontsize=6, class_names=["no", "si"], feature_names=nombre_columnas)
plt.show()
```

## Parametrizando el modelo de √°rbol

```python
modelo_arbol = DecisionTreeClassifier(max_depth=3, random_state=47)
modelo_arbol.fit(X_train, y_train)
```

```python
modelo_arbol.score(X_test, y_test)
```

```python
plt.figure(figsize=(25,10))
plot_tree(modelo_arbol, filled=True, fontsize=10, class_names=["no", "si"], feature_names=nombre_columnas)
plt.show()
```

## Normalizando las variables explicativas

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
normalizacion = MinMaxScaler()
X_train_normalizado = normalizacion.fit_transform(X_train)
```

## Uso del Modelo KNN (o algoritmo KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train_normalizado, y_train)
```

```python
#Para probar el modelo KNN, es necesario normalizar X_test
X_test_normalizado = normalizacion.fit_transform(X_test)
```

```python
#Evalua la performance del modelo
knn.score(X_test_normalizado, y_test)
```
